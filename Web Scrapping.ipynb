```python
import requests
from bs4 import BeautifulSoup
import csv
   
URL = "https://www.weightlossresources.co.uk/calories/calorie_counter/vegetables.htm"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
print(soup)
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
tables = soup.find_all('table')
```

```python
table = soup.find_all('table', class_='col_50_50')
```

```python
df = pd.DataFrame(columns=['Calories', 'Carbohydrates', 'Protiens', 'Fats', 'Fibre'])
```

```python
df
```

```python
for row in table:
    columns=row.tbody.find_all('tr')
    if(columns != []):
        Calories = columns[0].td.get_text()
        Carbohydrates = columns[1].td.get_text()
        Protiens = columns[2].td.get_text()
        Fats = columns[3].td.get_text()
        Fibre = columns[4].td.get_text()
        df = df.append({'Calories': Calories,  'Carbohydrates': Carbohydrates, 'Protiens': Protiens, 'Fats': Fats, 'Fibre': Fibre}, ignore_index=True)
```

```python
df
```

```python
name=name.reindex()
```

```python
name.index=arr
```

```python
name
```

```python
name.drop([7,33,75,74,72], axis=0, inplace=True)
```

```python
result = pd.concat([name, df], axis=1, join="inner")
```

```python
result
```

```python
result.to_csv("vegetables.csv",index=False)
```
