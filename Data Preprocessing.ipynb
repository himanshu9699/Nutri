```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import re
```

```python
df=pd.read_csv("Vegetables.csv")
```

```python
df
```

```python
for i in range(0,70):
    if 'Boiled' in df['Vegetables'][i]:
        df.drop([i],inplace=True)
```

```python
arr=[]
for i in range(0,56):
    arr.append(i)
```

```python
df.index=arr
```

```python
df
```

```python
for i in range(0,56):
    a=df['Vegetables'][i]
    df['Vegetables'][i]=a.split(',',1)[0]
```

```python
df.drop_duplicates(subset='Vegetables',inplace=True, keep="last")
```

```python
df
```

```python
df.shape
```

```python
arr=[]
for i in range(0,42):
    arr.append(i)
```

```python
df.index=arr
```

```python
df
```

```python
for i in range(0,42):
    df['Calories'][i]=re.sub("\D", "", df['Calories'][i])
    df['Carbohydrates'][i]=(re.findall(r'\d+\.\d+', str(df['Carbohydrates'][i])))
#     df['Carbohydrates'][i]=re.sub("\D", "", df['Carbohydrates'][i]) 
    df['Protiens'][i]=re.sub("\D", "", df['Protiens'][i])
    df['Fats'][i]=re.sub("\D", "", df['Fats'][i])
    df['Fibre'][i]=re.sub("\D", "", df['Fibre'][i])
```

```python
df
```

```python
df.to_csv("Vegetables_after_preprocessing.csv",index=False)
```
